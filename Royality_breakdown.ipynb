```python
import plotly.graph_objects as go

# Data
names = [
    "Arpad Dynasty",
    "House of Manfredi",
    "Kohanim",
    "House of Gontaut",
    "Clan Kinnaird",
    "Scottish Royalty",
    "House of Curiel",
    "House Percival",
    "Grand Dukes of Lithuania",
    "Persian Royalty",
    "Zhou Dynasty"
]

percentages = [
    39.4, 38.6, 13.2, 3.33, 1.95, 1.15, 0.67, 0.62, 0.38, 0.27, 0.24
]

# Location mapping
locations = {
    "Arpad Dynasty": "Hungary",
    "House of Manfredi": "Sicily, Italy",
    "Kohanim": "Various (Jewish diaspora)",
    "House of Gontaut": "Gascony, France",
    "Clan Kinnaird": "Perth and Kinross, Scotland",
    "Scottish Royalty": "Scotland",
    "House of Curiel": "Spain (Sephardic Jewish origin)",
    "House Percival": "England (Arthurian legends)",
    "Grand Dukes of Lithuania": "Lithuania",
    "Persian Royalty": "Persia (modern-day Iran)",
    "Zhou Dynasty": "China"
}

# Create the figure
fig = go.Figure()

# Add horizontal bars
fig.add_trace(go.Bar(
    y=names,
    x=percentages,
    orientation='h',
    text=[f'{p:.1f}% ({locations[n]})' for p, n in zip(percentages, names)],
    textposition='outside',
    hoverinfo='text',
    hovertext=[f'{name}<br>Percentage: {percentage:.2f}%<br>Location: {locations[name]}' for name, percentage in zip(names, percentages)]
))

# Customize the layout
fig.update_layout(
    title='Royality Breakdown by Y DNA',
    xaxis_title='Percentage',
    yaxis_title='Dynasty/Family',
    height=600,
    width=800,
    bargap=0.2,
    bargroupgap=0.1,
    yaxis={'categoryorder': 'total ascending'},
    uniformtext_minsize=8,
    uniformtext_mode='hide',
    xaxis_range=[0, 60]  # Set x-axis range from 0 to 50
)

# Show the figure
fig.show()
```
